```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import networkx as nx
import powerlaw
```

```python
def makeNetworkFromFiles(fedges, fnodes, delimiter):
    G = nx.Graph()
    fn=pd.read_csv(fnodes,header=None,delimiter=delimiter,index_col=0).transpose()
    n_attr=len(fn.index.values)
    attr=fn.index.values
    for n in fn.columns:
        attr_node=dict(zip(attr, fn[n].values))
        G.add_node(n,attr_dic=attr_node)
    fl=pd.read_csv(fedges,header=None,delimiter=delimiter)
    fl.columns = map(str, fl.columns)
    for L in fl.index.values:
        if fl.iloc[L,0] == fl.iloc[L,1]:
            # Piazza: "Self-loops affect many of the network properties 
            # that the problem set asks you to analyze, so I would
            # recommend removing these links for now."
            continue
        G.add_edge(fl.iloc[L,0],fl.iloc[L,1],weight=fl.iloc[L,3])
    GL = max(nx.connected_component_subgraphs(G), key=len)
    return GL

def remove_values_from_list(the_list, val):
    return [value for value in the_list if value != val]
```

```python
G = makeNetworkFromFiles('THE_LINKS-1.txt', 'THE_NODES-1.txt',' ')# csv is comma-delimited
k = dict(G.degree()).values()
k_weighted = dict(G.degree(weight='weight')).values()
print 'Sanity check: k min = ', min(k)
```

## 1. Calculate the degree distribution (make a plot similar to the inset of Fig. 1B for airport networks of the paper seen in class) (2 pts)

```python
fig, ax = plt.subplots()
fig.set_size_inches((9, 7))

n_bins = 20

k_logBins = np.logspace(np.log10(min(k)), np.log10(max(k)),num=n_bins)
k_logBinDensity, k_binedges = np.histogram(k, bins=k_logBins, density=True)

ax.loglog(k_logBins[:-1],k_logBinDensity,'o', markersize=10)

ax.set_xlabel('unweighted degree, $k$',fontsize=20)
ax.set_ylabel('$P(k)$',fontsize=20) 
plt.grid()
plt.grid(which='minor', linestyle='--')
plt.show()
```

## 2. Calculate the degree distribution of degree strengths or weighted degree (main plot of Figure 1B below of the paper seen in class) (2pts)

```python
fig, ax = plt.subplots()
fig.set_size_inches((9, 7))

n_bins = 20

k_logBins = np.logspace(np.log10(min(k_weighted)+0.001), np.log10(max(k_weighted)),num=n_bins)
k_logBinDensity, k_binedges = np.histogram(k_weighted, bins=k_logBins, density=True)


ax.loglog(k_logBins[:-1],k_logBinDensity,'o', markersize=10)

ax.set_xlabel('weighted degree, $w$',fontsize=20)
ax.set_ylabel('$P(w)$',fontsize=20)
plt.grid()
plt.grid(which='minor', linestyle='--')
plt.show()
```

## 3. Test the fit of a power law function for the degree distribution: $P(k) = C k^{-γ}$ What are the values of C and γ for the Power Law fit? Use the power law library (seen in Lecture 3 of unit 3) (3pts)

```python
fit=powerlaw.Fit(k,discrete=True)
```

```python
print 'γ =',fit.power_law.alpha
print 'C =',(fit.power_law.alpha - 1)*fit.xmin**(fit.power_law.alpha-1)
```
